```python
import keras
import numpy as np
from keras.models import load_model
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.add(keras.layers.Dense(units=1,input_shape=[1]))
model.compile(optimizer="sgd", loss="huber_loss")
```

```python

function = lambda x: x * (9 / 5) + 32

data = np.random.randint(-100, 100, size=3000)
train_percentage = 0.8
num_train = int(len(data) * train_percentage)

indices = np.random.permutation(len(data))
train_indices, test_indices = np.split(indices, [num_train])

x_train, x_test = data[train_indices], data[test_indices]
y_train, y_test = function(x_train), function(x_test)
```

```python
xs_train, xs_test = np.array(x_train, dtype=float), np.array(x_test, dtype=float)
ys_train, ys_test = np.array(y_train, dtype=float), np.array(y_test, dtype=float)
```

```python
fig, ax = plt.subplots()

table_data = [["Index", "a", "b"]]
table_data.extend([[i, a_i, b_i] for i, (a_i, b_i) in enumerate(zip(x_train, y_train))])

table = ax.table(cellText=table_data, loc="center", cellLoc="center", colLabels=None)
table.auto_set_font_size(False)
table.set_fontsize(8)

ax.axis("off")

plt.show()
```

```python
tf.test.is_gpu_available()
```

```python
model.fit(xs_train, ys_train, epochs=150)
```

```python
model.save("numberPattern.h5")
```

```python
loss = model.evaluate(xs_test, ys_test)
print(f"Test Loss: {loss}")
```

```python
loaded_model = load_model("numberPattern.h5")
new_data = np.array([5.0, 6.0, 7.0, 20, 40], dtype=float)
predictions = loaded_model.predict(new_data)

print("Predictions:")
print(predictions)
```
